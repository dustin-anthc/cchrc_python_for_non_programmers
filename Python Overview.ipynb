## Statistical Libraries

```python
# Sci-kit learn is a machine library that does pretty much anything statistical
import sklearn.linear_model as lm

# Initialize the model
regression_model = lm.LinearRegression()
```

```python
# Get our inputs to the model
X = aris[['CombinedFloorArea']]
y = aris[['Ann_Energy_Use_MMBTU']]

# Fit the model
regression_model.fit(X, y)
```

```python
# Get the R2 value to see if the floor area is a good predictor of energy use
regression_model.score(X, y)
```

```python
# import graphing library
import matplotlib.pyplot as plt

plt.scatter(X, y, color='g')
plt.plot(X, regression_model.predict(X), color='k')
plt.show()
```

```python
# Create a new model
multiple_regression_model = lm.LinearRegression()

# This time let's do a multiple regression, looking at the building size, AkWarm Rating Points, and Region
X = aris[['CombinedFloorArea', 'RatingPoints']]
y = aris[['Ann_Energy_Use_MMBTU']]

# Fit the model, this time also using the AkWarm Rating Points
multiple_regression_model.fit(X, y)
```

```python
multiple_regression_model.score(X, y)
```

### We have a stronger correlation, but these two variables still aren't a great predictor of modeled energy usage.

### Example: Cleaning Data

```python
# Read in a csv file
as_is_df = pd.read_csv(r"C:\Users\dustin\Google Drive\2017_Housing_Assessment\Data\Step_1_csv_files\CSV Housing Assessment - BRD - As_Is - 04-21-2016.csv",
                 index_col=None, header=0, parse_dates=['Rating_Date'])

# Sort by ProjectID
as_is_df = as_is_df.sort_values(by='ProjectID', ascending=True, axis=0)

# Display first five rows
as_is_df.head()
```

```python
# Do the same for the rest of the housing assessment data

post_df = pd.read_csv(r"C:\Users\dustin\Google Drive\2017_Housing_Assessment\Data\Step_1_csv_files\CSV Housing Assessment - BRD - Post_Improvement - 04-21-2016.csv",
                 index_col=None, header=0, parse_dates=['Rating_Date'])
post_df = post_df.sort_values(by='ProjectID', ascending=True, axis=0)

wx_as_is_df = pd.read_csv(r"C:\Users\dustin\Google Drive\2017_Housing_Assessment\Data\Step_1_csv_files\CSV Housing Assessment - BRD - Wx_As_Is - 04-21-2016.csv",
                 index_col=None, header=0, parse_dates=['Rating_Date'])
wx_as_is_df = wx_as_is_df.sort_values(by='ProjectID', ascending=True, axis=0)

df_list = [as_is_df, post_df, wx_as_is_df]
```

```python
def clean_dataframe(df):
    # Query dataframe for only single family or mobile home
    single_fam_or_mobile = df.query("HouseType == 'Single_Family' or HouseType == 'Mobile_Home'")
    
    # Create fullname column for deduplication purposes
    single_fam_or_mobile['FullName'] = single_fam_or_mobile.ClientFirst.str.lower() + ' ' \
                                       + single_fam_or_mobile.ClientLast.str.lower()
    
    single_fam_or_mobile['City'] = single_fam_or_mobile.City.str.lower()
    single_fam_or_mobile['Address'] = single_fam_or_mobile.Address.str.lower()
    
    # Because the dataframes are sorted by ProjectID, setting the take_last option to True means that the file with the most
    # recent ProjectID will be kept.
    
    # file name, address, winwall ratio
    cleaned_df = single_fam_or_mobile.drop_duplicates(subset=('Address', 'FileName' ,'WinWallRatio', 'City',
                                                    'RatingType', 'HouseType'), keep='last')

    # If filename, address, and concatenated name are the same, dedup
    cleaned_df = cleaned_df.drop_duplicates(subset=('Address', 'FileName' , 'FullName', 'City','RatingType', 
                                                    'HouseType'), keep='last')

    # concat name, file name, winwallratio, dedup
    cleaned_df = cleaned_df.drop_duplicates(subset=('FileName' , 'FullName', 'WinWallRatio', 'City',
                                                    'RatingType', 'HouseType'), keep='last')

    # if Rating Points, Concatenated name, floor expression, and WinWallRatio are the same, dedup
    cleaned_df = cleaned_df.drop_duplicates(subset=('RatingPoints', 'FullName', 'FloorExpression', 'WinWallRatio', 'City',
                                                    'RatingType', 'HouseType'), keep='last')

    # if Address, Concatenated name, and FloorExpression are the same, dedup
    cleaned_df = cleaned_df.drop_duplicates(subset=('Address', 'FullName', 'FloorExpression', 'City',
                                                    'RatingType', 'HouseType'), keep='last')

    # if FileName, Concatenated name, and FloorExpression are the same, dedup
    cleaned_df = cleaned_df.drop_duplicates(subset=('FileName', 'FullName', 'FloorExpression', 'City',
                                                    'RatingType', 'HouseType'), keep='last')

    # if FullName, FileName, and City are the same, dedup
    cleaned_df = cleaned_df.drop_duplicates(subset=('FileName', 'FullName', 'City',
                                                    'RatingType', 'HouseType'), keep='last')
    
    # Create CSV file with deduplicated results
    file_location = "C:\\Users\\dustin\\Google Drive\\Python Training\\"
    file_name = cleaned_df.RatingType[0] + "_double_deduped_mislabeled_fixed.csv"
    output_path = os.path.join(file_location, file_name)
    cleaned_df.to_csv(output_path)
    
    return cleaned_df
```

```python
for df in df_list:
    print "The total number of original AkWarm ratings of the %s type is %d" % (df.RatingType[0], df.shape[0])
    cleaned_df = clean_dataframe(df)
    print "%d ratings were cleaned from this dataset.  You're welcome. '\n' " % (df.shape[0] - cleaned_df.shape[0])
```
